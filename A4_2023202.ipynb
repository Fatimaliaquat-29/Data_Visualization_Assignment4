# Data Structures Assignment 4: Tree and Graph Visualizations

## Dataset 1: Flare (Tree Visualization)
* **Source**: The Vega datasets repository. We are using a local `flare.json` file representing the dataset.
* **Nodes Represent**: Software packages, classes, and functions within the Flare visualization library.
* **Edges Represent**: The parent-child hierarchical relationship (e.g., a package containing a class).

## Dataset 2: Zachary's Karate Club (Graph Visualization)
* **Source**: Built-in dataset from the Python `networkx` library, originally observed by Wayne Zachary (1977).
* **Nodes Represent**: 34 individual members of a university karate club.
* **Edges Represent**: Social interactions and friendships between members outside the club context.

```python
import networkx as nx
import matplotlib.pyplot as plt
import json
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
plt.style.use("dark_background")
```

## 1. Flare Dataset (Tree Structure)
We load the hierarchical data from the local `flare.json` file and construct a directed graph (`DiGraph`).

```python
# Load local JSON data
with open('flare.json', 'r') as f:
    flare_data = json.load(f)

flare_tree = nx.DiGraph()

# Build the tree from flat JSON list (id, name, parent)
for item in flare_data:
    node_id = item.get('id')
    node_name = item.get('name', 'Unknown')
    flare_tree.add_node(node_id, label=node_name)

for item in flare_data:
    node_id = item.get('id')
    if 'parent' in item and item['parent'] is not None:
        parent_id = item['parent']
        flare_tree.add_edge(parent_id, node_id)

print(f"Flare Tree constructed with {flare_tree.number_of_nodes()} nodes and {flare_tree.number_of_edges()} edges.")
```

### Layout 1: Kamada-Kawai
The Kamada-Kawai force-directed layout is very good at clustering child nodes around their parents.

```python
plt.figure(figsize=(16, 16), facecolor='#0a0a0b')
ax = plt.gca()
ax.set_facecolor("#0a0a0b")
plt.title("Flare Dataset - Kamada-Kawai Layout", fontsize=18, color="white")

pos_flare_kk = nx.kamada_kawai_layout(flare_tree)
nx.draw(flare_tree, pos_flare_kk, 
        node_size=10, 
        node_color="#1f78b4", 
        edge_color="#444444", 
        width=0.3, 
        arrows=False, 
        with_labels=False, 
        alpha=0.6)

plt.savefig('visualizations/flare_kamada_kawai.png', dpi=300, bbox_inches='tight', facecolor='#0a0a0b', transparent=False)
plt.close()
```

### Layout 2: Spring Layout
The Spring layout (Fruchterman-Reingold force-directed algorithm) gives another organic look to the tree structure.

```python
plt.figure(figsize=(16, 16), facecolor='#0a0a0b')
ax = plt.gca()
ax.set_facecolor("#0a0a0b")
plt.title("Flare Dataset - Spring Layout", fontsize=18, color="white")

pos_flare_spring = nx.spring_layout(flare_tree, k=0.15, iterations=50)
nx.draw(flare_tree, pos_flare_spring, 
        node_size=10, 
        node_color="#33a02c", 
        edge_color="#444444", 
        width=0.3, 
        arrows=False, 
        with_labels=False, 
        alpha=0.6)

plt.savefig('visualizations/flare_spring.png', dpi=300, bbox_inches='tight', facecolor='#0a0a0b', transparent=False)
plt.close()
```

## 2. Zachary's Karate Club (Graph Structure)
This dataset is built into NetworkX. We will visualize it utilizing the `club` attribute to color the nodes, which represent the two separate factions that emerged.

```python
karate_graph = nx.karate_club_graph()
print(f"Karate Club Graph loaded with {karate_graph.number_of_nodes()} nodes and {karate_graph.number_of_edges()} edges.")

# Assign colors based on the 'club' attribute (Mr. Hi vs Officer)
club_colors = ['#1f78b4' if karate_graph.nodes[node]['club'] == 'Mr. Hi' else '#e31a1c' for node in karate_graph.nodes()]
node_labels = {node: str(node) for node in karate_graph.nodes()}
```

### Layout 1: Circular Layout
The Circular layout places nodes uniformly on a circle, which is great for showing edge connections without node overlaps.

```python
plt.figure(figsize=(12, 12), facecolor='#0a0a0b')
ax = plt.gca()
ax.set_facecolor("#0a0a0b")
plt.title("Karate Club - Circular Layout", fontsize=18, color="white")

pos_karate_circ = nx.circular_layout(karate_graph)
nx.draw(karate_graph, pos_karate_circ, 
        labels=node_labels,
        node_color=club_colors, 
        node_size=400, 
        font_size=9, 
        font_weight='bold', 
        font_color='white',
        edge_color="#777777",
        width=0.8)

plt.savefig('visualizations/karate_circular.png', dpi=300, bbox_inches='tight', facecolor='#0a0a0b', transparent=False)
plt.close()
```

### Layout 2: Spring Layout
The Spring layout pushes connected nodes closer together, naturally revealing the two distinct community clusters.

```python
plt.figure(figsize=(12, 12), facecolor='#0a0a0b')
ax = plt.gca()
ax.set_facecolor("#0a0a0b")
plt.title("Karate Club - Spring Layout", fontsize=18, color="white")

pos_karate_spring = nx.spring_layout(karate_graph, k=0.3, seed=42) # Set seed for reproducible layout
nx.draw(karate_graph, pos_karate_spring, 
        labels=node_labels,
        node_color=club_colors, 
        node_size=400, 
        font_size=9, 
        font_weight='bold',
        font_color='white',
        edge_color="#777777",
        width=0.8)

plt.savefig('visualizations/karate_spring.png', dpi=300, bbox_inches='tight', facecolor='#0a0a0b', transparent=False)
plt.close()
```
